```python
corr_indo = indo[
[
"Rating",
"Total Reviewers",
"log_review"
]
].corr()

corr_indo
```

```python
indo["popularity_score"] = (
    (
        indo["Rating"]*0.7
    )
    +
    (
        (
            indo["log_review"]
            /
            indo["log_review"].max()
        )*5
    )*0.3
)
```

```python
top_products = (
    indo
    .sort_values(
        "popularity_score",
        ascending=False
    )
)

top_products[
[
"Brand",
"Name",
"Rating",
"Total Reviewers",
"popularity_score"
]
].head(20)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

indo["rating_scaled"] = (
    scaler.fit_transform(
        indo[
            ["Rating"]
        ]
    )
)

indo["review_scaled"] = (
    scaler.fit_transform(
        indo[
            ["log_review"]
        ]
    )
)

indo["recommendation_score"] = (
    (
        indo[
            "rating_scaled"
        ]*0.6
    )
    +
    (
        indo[
            "review_scaled"
        ]*0.4
    )
)
```

```python
indo[
[
"Brand",
"Name",
"Rating",
"Total Reviewers",
"recommendation_score"
]
].sort_values(
"recommendation_score",
ascending=False
).head(20)
```
